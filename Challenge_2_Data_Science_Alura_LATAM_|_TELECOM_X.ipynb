**Extracci贸n**:

```python
import pandas as pd       #importaci贸n de la biblioteca pandas 'as pd' significa que ahora se llama as铆
datos = pd.read_json('/content/TelecomX_Data.json') #importaci贸n de la API(archivo descargado previamente .json)<<<< .read_json
datos.head()
```

```python
datos.info()  #verificamos si era Data frame o si hay nulos
datos.columns #verificamos las columnas
```

**Transformaci贸n:**

```python
# Desanidar cada columna tipo dict
customer_df = pd.json_normalize(datos['customer'])
phone_df    = pd.json_normalize(datos['phone'])
internet_df = pd.json_normalize(datos['internet'])
account_df  = pd.json_normalize(datos['account'])

# Concatenar todo en un solo DataFrame
datos = pd.concat([customer_df, phone_df, internet_df, account_df], axis=1)

# Verificar el resultado
datos.head()
```

```python
datos.isnull().sum().sort_values(ascending=False) #Verificamos que no haya valores nulos
```
