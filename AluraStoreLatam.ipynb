### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda_1 = tienda['Precio'].sum()
print("\nIngreso Total Tienda 1:",ingreso_tienda_1)
ingreso_tienda_2 = tienda2['Precio'].sum()
print("\nIngreso Total Tienda 2:",ingreso_tienda_2)
ingreso_tienda_3 = tienda3['Precio'].sum()
print("\nIngreso Total Tienda 3:",ingreso_tienda_3)
ingreso_tienda_4 = tienda4['Precio'].sum()
print("\nIngreso Total Tienda 4:",ingreso_tienda_4)

```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

ingresos = [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(15, 10))
plt.bar(tiendas, ingresos, color='blue')
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tiendas")
plt.yticks([1150880, 1116343, 1098019, 1038375])
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
categorias_tienda_1 = tienda['Categoría del Producto'].value_counts()
print("\nTienda 1 - Categorías más vendidas:\n", categorias_tienda_1.head(3))
categorias_tienda_2 = tienda2['Categoría del Producto'].value_counts()
print("\nTienda 2 - Categorías más vendidas:\n", categorias_tienda_2.head(3))
categorias_tienda_3 = tienda3['Categoría del Producto'].value_counts()
print("\nTienda 3 - Categorías más vendidas:\n", categorias_tienda_3.head(3))
categorias_tienda_4 = tienda4['Categoría del Producto'].value_counts()
print("\nTienda 4 - Categorías más vendidas:\n", categorias_tienda_4.head(3))
```

```python
categorias_tienda_1.head(3).plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title("Categorías Más Vendidas - Tienda 1")
plt.ylabel('')
plt.show()
categorias_tienda_2.head(3).plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title("Categorías Más Vendidas - Tienda 2")
plt.ylabel('')
plt.show()
categorias_tienda_3.head(3).plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title("Categorías Más Vendidas - Tienda 3")
plt.ylabel('')
plt.show()
categorias_tienda_4.head(3).plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title("Categorías Más Vendidas - Tienda 4")
plt.ylabel('')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda_1 = tienda['Calificación'].mean()
print("\nTienda 1 - Calificación:\n", round(calificacion_tienda_1,2))
calificacion_tienda_2 = tienda2['Calificación'].mean()
print("\nTienda 2 - Calificación:\n", round(calificacion_tienda_2,2))
calificacion_tienda_3 = tienda3['Calificación'].mean()
print("\nTienda 3 - Calificación:\n", round(calificacion_tienda_3,2))
calificacion_tienda_4 = tienda4['Calificación'].mean()
print("\nTienda 4 - Calificación:\n", round(calificacion_tienda_4,2))
```

# 4. Productos más y menos vendidos

```python
producto_tienda_1 = tienda['Producto'].value_counts()
print("\nMás vendido T1 -",producto_tienda_1.idxmax(),"  #  ",producto_tienda_1.max())
print("Menos vendido T1 -",producto_tienda_1.idxmin(),"  #  ",producto_tienda_1.min())
producto_tienda_2 = tienda2['Producto'].value_counts()
print("\nMás vendido T2 -",producto_tienda_2.idxmax(),"  #  ",producto_tienda_2.max())
print("Menos vendido T2 -",producto_tienda_2.idxmin(),"  #  ",producto_tienda_2.min())
producto_tienda_3 = tienda3['Producto'].value_counts()
print("\nMás vendido T3 -",producto_tienda_3.idxmax(),"  #  ",producto_tienda_3.max())
print("Menos vendido T3 -",producto_tienda_3.idxmin(),"  #  ",producto_tienda_3.min())
producto_tienda_4 = tienda4['Producto'].value_counts()
print("\nMás vendido T4 -",producto_tienda_4.idxmax(),"  #  ",producto_tienda_4.max())
print("Menos vendido T4 -",producto_tienda_4.idxmin(),"  #  ",producto_tienda_4.min())
```

# 5. Envío promedio por tienda

```python
Envio_tienda_1 = tienda['Costo de envío'].mean()
print("\nTienda 1:\n", round(Envio_tienda_1))
Envio_tienda_2 = tienda2['Costo de envío'].mean()
print("\nTienda 2:\n", round(Envio_tienda_2))
Envio_tienda_3 = tienda3['Costo de envío'].mean()
print("\nTienda 3:\n", round(Envio_tienda_3))
Envio_tienda_4 = tienda4['Costo de envío'].mean()
print("\nTienda 4:\n", round(Envio_tienda_4))
```

```python
calificaciones = [calificacion_tienda_1, calificacion_tienda_2, calificacion_tienda_3, calificacion_tienda_4]
costos_envio = [Envio_tienda_1, Envio_tienda_2, Envio_tienda_3, Envio_tienda_4]

plt.figure(figsize=(8, 5))
plt.scatter(costos_envio, calificaciones, color='blue')

for i, txt in enumerate(tiendas):
    plt.annotate(txt, (costos_envio[i], calificaciones[i]))

plt.title("Relación entre Calificación y Costo de Envío")
plt.xlabel("Costo de Envío Promedio ($)")
plt.ylabel("Calificación Promedio")
plt.grid(True)
plt.tight_layout()
plt.show()
```
